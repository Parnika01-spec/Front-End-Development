```python
# BASIC DATA STRUCTURE IN PYTHON

# Strings
#Ordered(Positions matter), Immutable(cannot be changed once created) Iterable(can loop through it)
#Why Immputable- since python saves string in a special memory area and reuses them to save the memory, changing breaks optimization
Name= "Parnika Verma"
print(Name[-2])
print(Name[-6])
print(Name[-8])
print(Name[0])

#List
#Ordered, Mutable(user can modify the values), alloes duplicated, store any data type

fruits= ["apple", "mango", "grapes"]
print(fruits)
fruits.append ("orange")
print(fruits)

#Tuple(Lists but immutable)
#Ordered, Immutable< fast(since the data is fixed), allows duplicates
#Why use tuples- When we want the data to be fixed; like the coordinatess, monthn of the year, etc.
colors= ("red", "black", "blue", "violet", "white")
print(colors)
print(colors[0])


#Sets- for storing unique data. Fast, mutable, unordered, doesn't allow duplicate data.
marks= {99, 100, 94, 95, 97, 99}
print(marks)
#Why use set?- when you want unique values or want to perform set operations.
```
